PREPROCESSING

```python
df.isnull().sum().sum()
```

```python
import numpy as np
from scipy import stats
z = np.abs(stats.zscore(df))
z
```

```python
len(np.where(z>3)[0])
```

```python
len(list(set(np.where(z>3)[0])))
```

```python
outliers = list(set(np.where(z>3)[0]))
```

```python
df = df.drop(outliers, axis = 0).reset_index(drop = True)
```

```python
df.shape
```

```python
X = df.iloc[:, 0:-1]
```

```python
X.shape
```

```python
y = df.iloc[:,-1]
```

```python
y.head()
```

```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler
X = MinMaxScaler().fit_transform(X)
```

```python
from sklearn.model_selection import train_test_split, cross_val_score, cross_val_predict
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 42)
```

MODELING

```python
#1 LOGISTIC REGRESSION
from sklearn.linear_model import LogisticRegression
loj = LogisticRegression(solver = "liblinear")
loj_model = loj.fit(X_train, y_train)
```

```python
from sklearn.model_selection import train_test_split, cross_validate
cv = cross_validate(loj_model,X_train,y_train, cv = 3, n_jobs=-1)
```

```python
cv["test_score"].mean()
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report
y_pred = loj_model.predict(X_test)
confusion_matrix(y_test, y_pred)
```

```python
print(classification_report(y_test, y_pred))
```

```python
#2 DECISION TREE
from sklearn.tree import DecisionTreeClassifier
tree_model = DecisionTreeClassifier().fit(X_train, y_train)
```

```python
tree_model.score(X_train, y_train)
```

```python
tree_model.score(X_test, y_test)
```

```python
y_pred = tree_model.predict(X_test)
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report

confusion_matrix(y_test, y_pred)
```

```python
y_pred.mean()
```

```python
print(classification_report(y_test, y_pred))
```

```python
from sklearn.model_selection import GridSearchCV

tree_params = {"min_samples_split": range(2,50),
               "max_depth": range(1,10)}
tree_cv = GridSearchCV(tree_model, tree_params, cv = 10, n_jobs = -1).fit(X_train, y_train)

tree_cv.best_params_
```

```python
tree_tuned = DecisionTreeClassifier(max_depth = 9, min_samples_split = 3).fit(X_train, y_train)
```

```python
tree_tuned.score(X_train, y_train)
```

```python
tree_tuned.score(X_test, y_test)
```

```python
y_pred = tree_tuned.predict(X_test)
confusion_matrix(y_test, y_pred)
```

```python
print(classification_report(y_test, y_pred))
```
